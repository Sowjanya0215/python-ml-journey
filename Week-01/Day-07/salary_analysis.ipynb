```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x=[1,2,3]
y=[4,5,6]
plt.plot(x,y)
```

```python
data={"salaries": [100000,30000,25000, 30000, 28000, 35000, 40000]}
df=pd.DataFrame(data)
plt.plot(df["salaries"])
plt.show()
```

```python
plt.plot(df["salaries"],color="blue",marker="o",linestyle="--")
plt.xlabel("months")
plt.ylabel("salary")
plt.grid()
```
